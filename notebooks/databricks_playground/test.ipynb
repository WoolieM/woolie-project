```python
from databricks.connect import DatabricksSession

# This 'wakes up' the bridge between your VM and the Cloud
spark = DatabricksSession.builder.getOrCreate()

# Verify the connection
print(f"✅ Connected to: {spark.client.host}")
```

```python
# This query runs on the Databricks Server, not your VM
df_cloud = spark.sql("SELECT current_user() as user, current_catalog() as catalog, current_timestamp() as cloud_time")
display(df_cloud)
```

```python
# 1. Create Dummy Data
data = [
    (1, "Woolie", "Data Engineer"),
    (2, "Cassandra", "Quality Assurance"),
    (3, "Melbourne_Bot", "Automation")
]

# 2. Define Schema and Create DataFrame
columns = ["emp_id", "emp_name", "role"]
df = spark.createDataFrame(data, schema=columns)

# 3. Write to a Delta Table (The 'Insert')
# This creates a persistent table in your 'default' database
df.write.format("delta").mode("overwrite").saveAsTable("default.woolie_employees_test")

# 4. Show the results (Use display() for a nice Databricks table view)
display(spark.table("default.woolie_employees_test"))
```
