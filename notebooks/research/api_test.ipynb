```python
import sys
import os
import asyncio
from pathlib import Path
from dotenv import load_dotenv

# 1. Path Handshake: Ensure notebook can find the 'app' directory
root_path = Path(os.getcwd()).parent.parent
if str(root_path) not in sys.path:
    sys.path.append(str(root_path))

# 2. Security: Load your real credentials
load_dotenv(root_path / ".env")
api_key = os.getenv("COINGECKO_API_KEY")

# 3. Import your updated Class
from app.simulation.clients.coingecko import CoinGeckoClient

# 4. Initialize the Real Handshake
client = CoinGeckoClient(api_key=api_key)

# 5. Fetch the expanded data (BTC/ETH in AUD/USD + Volume)
# We use the updated method name 'get_prices'
data = await client.get_prices(
    coin_ids="bitcoin,ethereum", 
    currencies="aud,usd"
)

# 6. Success! Display the results
print("--- RAW API RESPONSE ---")
print(data)

print("\n--- SENIOR DATA VIEW ---")
# Accessing your 3 requested data points + volumes
btc_aud = data['bitcoin']['aud']
btc_usd = data['bitcoin']['usd']
eth_aud = data['ethereum']['aud']

print(f"BTC Price (Sydney): ${btc_aud:,.2f} AUD | Vol: ${data['bitcoin']['aud_24h_vol']:,.0f}")
print(f"BTC Price (Global): ${btc_usd:,.2f} USD | Vol: ${data['bitcoin']['usd_24h_vol']:,.0f}")
print(f"ETH Price (Sydney): ${eth_aud:,.2f} AUD | Vol: ${data['ethereum']['aud_24h_vol']:,.0f}")
print(f"Captured at (Unix): {data['bitcoin']['last_updated_at']}")
```
